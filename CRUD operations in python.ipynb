```python
#Nombre: Aplicacion CRUD con conexion a MySQL
#Objetivo:Hacer una aplicacion con interfaz grafica que haga las operaciones  principales(CRUD) con conexion a una base de 
#datos en mySQL
#Autor: Marcos Oswaldo Gil Naranjo
#Fecha:15/jul/2019 


import tkinter as tk
from tkinter import *
from tkinter import messagebox
import pymysql
from PIL import Image, ImageDraw
from reportlab.pdfgen import canvas

class base:
    
    def __init__(self,wv):
         # Modificamos el tama単o de la ventana
        wv.geometry("300x200")
        # Titulo de la ventana
        wv.title("Base de datos ")
        #Creamos los botones
        bag = Button(None, text='Agregar', command=self.Agregar)
        bb = Button(None, text='Borrar', command=self.Borrar)
        bac = Button(None, text='Actualizar', command=self.Actualizar)
        br = Button(None, text='Reporte', command=self.Reporte)

        # Botones
        bag.place(x = 10, y=75)
        bb.place(x = 75, y=75)
        bac.place(x = 130, y=75)
        br.place(x = 200, y=75)
        

    def Agregar(self):
        wv2 = tk.Tk()
        wv2.geometry("300x200")
        wv2.title("Base de datos - Agregar ")
        l1 = Label(wv2, text="Clave")
        l1.place(x=10,y=10)
        entry1 = tk.Entry(wv2)
        entry1.focus()
        entry1.place(x=100, y=10)
        l2 = Label(wv2, text="Nombre")
        l2.place(x=10,y=40)
        entry2 = tk.Entry(wv2)
        entry2.place(x=100, y=40)
        bag = Button(wv2, text='Agregar', command= lambda:self.AgregarSQL(entry1.get(),entry2.get())).place(x = 10, y=60)
        wv2.mainloop()

    def AgregarSQL(self,c,n):
        sql = "Insert into keyword values ("+c+",'"+n+"');"
        try:
           # Execute the SQL command
           cursor.execute(sql) 
           # Commit your changes in the database
           db.commit()
        except:
           # Rollback in case there is any error
           db.rollback()
           self.vError()
           return
        self.vCorrecto()
    
    def Borrar(self):
        wv3 = tk.Tk()
        # Modificamos el tama単o de la ventana
        wv3.geometry("300x200")
        # Titulo de la ventana
        wv3.title("Base de datos - Borrar ")
        l1 = Label(wv3, text="Clave")
        l1.place(x=10,y=10)
        entry1 = tk.Entry(wv3)
        entry1.focus()
        entry1.place(x=100, y=10)
        bag = Button(wv3, text='Borrar', command= lambda:self.BorrarSQL(entry1.get())).place(x = 10, y=60)
        wv3.mainloop()

    def BorrarSQL(self,n):
        sql = "delete from keyword where clave="+n
        try:
           # Execute the SQL command
           cursor.execute(sql) 
           # Commit your changes in the database
           db.commit()
        except:
           # Rollback in case there is any error
           db.rollback()
           self.vError()
           return
        self.vCorrecto()
    
    def Actualizar(self):
        wv4 = tk.Tk()
        # Modificamos el tama単o de la ventana
        wv4.geometry("200x150")
        # Titulo de la ventana
        wv4.title("Base de datos - Actualizar ")
        l1 = Label(wv4, text="Clave")
        l1.place(x=10,y=10)
        entry1 = tk.Entry(wv4)
        entry1.focus()
        entry1.place(x=100, y=10)
        l2 = Label(wv4, text="Nombre")
        l2.place(x=10,y=40)
        entry2 = tk.Entry(wv4)
        entry2.place(x=100, y=40)
        bag = Button(wv4, text='Actualizar', command= lambda:self.ActualizarSQL(entry1.get(),entry2.get())).place(x = 10, y=60)
        wv4.mainloop()

    def ActualizarSQL(self,c,n):
        sql = "update keyword set nombre='"+n+"'  where clave="+c+ "; "
        try:
           # Execute the SQL command
           cursor.execute(sql) 
           # Commit your changes in the database
           db.commit()
        except:
           # Rollback in case there is any error
           db.rollback()
           self.vError()
           return
        self.vCorrecto()
    
    def Reporte(self):
        #bUSCAR
        # Creamos las ventana
        wv1 = tk.Tk()
        # Modificamos el tama単o de la ventana
        wv1.geometry("500x400")
        # Titulo de la ventana
        wv1.title("Base de datos - Reporte ")

        sql = "Select * from keyword "

        cursor.execute(sql)

        results = cursor.fetchall()
        for row in results:
            clave=row[0]
            nombre=row[1]
            w = Label(wv1, text="clave = {0}, nombre = {1}".format(clave,nombre))
            w.pack()
        bpdf = Button(wv1, text='Guardar en PDF', command=lambda:self.crearPDF(results))
        bimg = Button(wv1, text='Guardar en Imagen', command=lambda:self.crearImg(results))
        bpdf.pack()
        bimg.pack()
            
       #Ventanas de dialogo 
    def vError(self):
        messagebox.showinfo(message="Error: Accion no realizada", title="Error")
        
    def vCorrecto(self):
        messagebox.showinfo(message="Accion realizada correctamente", title="Correcto")
        
#metodo para crear un archivo pdf        
    def crearPDF(self,results):
        c = canvas.Canvas("Reporte.pdf")
        text = c.beginText(50, 800)
        for row in results:
            clave=row[0]
            nombre=row[1]
            text.textLine("clave = {0}, nombre = {1}".format(clave,nombre))
        c.drawText(text)
        c.save()
        self.vCorrecto()
# metodo para crear una imagen png        
    def crearImg(self, results):
        img = Image.new('RGB', (800, 500), color=(80, 99, 250))
 
        d = ImageDraw.Draw(img)
        x=10
        for row in results:
            clave=row[0]
            nombre=row[1]
            x=x+15
            d.text((10,x),"clave = {0}, nombre = {1}".format(clave,nombre))
        
        img.save('Reporte.png')
        self.vCorrecto()

def main():
    wv = tk.Tk()
    ventana=base(wv)
        
if __name__ =="__main__":
    try:
        #cadenad de conexion al servidor de la base de datos
        db = pymysql.connect("localhost","root","","automatasii")
        cursor = db.cursor()
        print("Conexion Establecida...")
    except:
        print("Conexion fallida...") 
    main()
    wv.mainloop()
```
